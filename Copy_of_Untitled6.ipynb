```python
#importing  libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive


# Mount Google Drive
drive.mount('/content/drive')

# Define CSV file path
csv_file_path_listings = '/content/drive/MyDrive/SPC-DF_POP_COAST-2.0-all_.csv'

# Create dataframe from CSV file
df = pd.read_csv(csv_file_path_listings)
df.head()
```

```python
# Exclude rows with 'PERCENT: percent' in the 'UNIT_MEASURE: Unit of measure' column
df = df[df['UNIT_MEASURE: Unit of measure'] != 'PERCENT: percent']

# Display the updated DataFrame
df.head()
```

```python
# Drop the specified columns
columns_to_drop = ['INDICATOR: Indicator', 'UNIT_MEASURE: Unit of measure']
df = df.drop(columns=columns_to_drop)

# Display the updated DataFrame
df.head()
```

```python
# Keep only rows with 'PG: Papua New Guinea' in the 'GEO_PICT: Pacific Island Countries and territories' column
df = df[df['GEO_PICT: Pacific Island Countries and territories'] == 'PG: Papua New Guinea']

# Display the updated DataFrame
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a histogram with seaborn
plt.figure(figsize=(12, 6))
```

```python
sns.histplot(data=df, x='TIME_PERIOD: Time', y='OBS_VALUE', hue='RANGE: Range', multiple='stack', stat='count', binwidth=1)
```

```python
# Set labels and title
plt.xlabel('Time Period')
plt.ylabel('OBS Value')
plt.title('Histogram of OBS Value over Time with Range Groups')

# Show the legend
plt.legend(title='Range', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```
